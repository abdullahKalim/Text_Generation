```python
df=df.drop(na_50,axis=1) #droping colums having more than 50% missing values
```

```python
from sklearn.impute import SimpleImputer
impute=SimpleImputer(strategy='most_frequent')
df_imp=impute.fit_transform(df)
```

```python
df_imp=pd.DataFrame(df_imp,columns=df.columns)
```

```python
df_imp[object_columns].info()
```

```python
from sklearn.preprocessing import OneHotEncoder
df_cat=df_imp[object_columns]
df_num=df_imp.drop(object_columns,axis=1)
encoder=OneHotEncoder(sparse=False,handle_unknown='ignore')
one_hot=encoder.fit_transform(df_cat)
```

```python
Y=df_num["hospital_death"]
df_num=df_num.drop("hospital_death",axis=1)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
df_num=scaler.fit_transform(df_num)
```

```python
df_new=np.concatenate((df_num,one_hot),axis=1)
```

```python
from sklearn.model_selection import train_test_split
train_X,test_X,train_y,test_y=train_test_split(df_new,Y,test_size=0.2,random_state=10)
train_X_df=pd.DataFrame(train_X)
train_y_df=pd.DataFrame(train_y)
```

```python
model=Sequential()
model.add(Dense(64,input_shape=(train_X.shape[-1],),activation='relu'))
model.add(Dense(32,activation='relu'))
model.add(Dense(16,activation='relu'))
model.add(Dense(8,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
# metrics that might be useful for present project
metrics = [
    keras.metrics.Precision(name="precision"),
    keras.metrics.Recall(name="recall"),
    keras.metrics.AUC(curve='ROC')
]


### Compiling the model
### choosing an appropirate optimiser and loss function
model.compile(optimizer=keras.optimizers.Adam(0.01),
              loss='binary_crossentropy',
              metrics=metrics)
```

```python
train_X=np.asarray(train_X).astype('float32')
```

```python
train_y=np.asarray(train_y).astype('float32')
```

```python
epochs=20
batch_size=256
history=model.fit(train_X,train_y,epochs=epochs,batch_size=batch_size,class_weight={0: 0.5, 1: 5})
```

```python
model_tuned=Sequential()
model_tuned.add(Dense(416,input_shape=(train_X.shape[-1],),activation='relu'))
model_tuned.add(Dense(416,activation='relu'))
model_tuned.add(Dense(1,activation='sigmoid'))
```

```python
model_tuned.compile(optimizer=keras.optimizers.Adam(0.0001),
              loss='binary_crossentropy',
              metrics=metrics)
```

```python
history_tuned=model_tuned.fit(train_X,train_y,epochs=epochs,batch_size=batch_size,class_weight={0: 0.5, 1: 5})
```
